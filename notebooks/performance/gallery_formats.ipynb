```python
# slug = "ephes_blog"
slug = "das_claas_und_nora_blog"
blog = Blog.objects.get(slug=slug)
```

```python
def get_gallery_blocks(raw_data):
    gallery_blocks = []
    for root_block in raw_data:
        for block in root_block["value"]:
            if block["type"] == "gallery":
                gallery_blocks.append(block)
    return gallery_blocks
```

```python
for post in blog.unfiltered_published_posts:
    if post.galleries.count() > 0:
        # rprint(post.body.raw_data)
        gallery_blocks = get_gallery_blocks(post.body.raw_data)
        rprint(gallery_blocks)
```
