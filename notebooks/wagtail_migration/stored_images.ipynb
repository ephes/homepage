# Delete Renditions

```python
from pathlib import Path
```

```python
complete = 0
rendition_dir = Path("images")
dirs, rendition_files = default_storage.listdir(str(rendition_dir))
for num, name in enumerate(rendition_files):
    rendition_path = str(rendition_dir / name)
    rendition_size = default_storage.size(rendition_path)
    complete += rendition_size
    print(rendition_path, rendition_size)
print(complete)
```

```python
rendition_dir = Path("images")
dirs, rendition_files = default_storage.listdir(str(rendition_dir))
for num, name in enumerate(rendition_files):
    rendition_path = rendition_dir / name
    # print(rendition_path)
    default_storage.delete(str(rendition_path))
    if num % 10 == 0:
        print(num)
    
# print(dirs)
# print(rendition_files)
```
